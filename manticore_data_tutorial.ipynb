## Import Required Libraries

````python
import numpy as np
import matplotlib.pyplot as plt
import h5py
from manticore_data import ManticoreDataLoader, ManticoreGeneration

# Set up matplotlib for better-looking plots
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
````

## Initialize the Data Loader

The `ManticoreDataLoader` connects to the S3 data repository and discovers available data products.

````python
# Initialize the loader
loader = ManticoreDataLoader()

# Check available products
#print("Available products:")
#products = loader.get_products(ManticoreGeneration.Local)
#for product in products:
#    print(f"  - {product}")
````

## Download Velocity and Density Field for MCMC Step 0

The velocity field file contains both density and momentum data. We'll download it without the caching option. There are a total of 80 field files (one for each Manticore-Local indepedent realization of the posterior).

**File structure:**
- `density`: Mass field [Msun/h]
- `p0`, `p1`, `p2`: Momentum components (x, y, z) [(Msun/h) (km/s)]
- To get velocity: divide momentum by `density` (velocity in km/s)

`h = 0.681`

````python
# Get velocity field files for resolution 1024
velocity_files = loader.get_velocity_field(
    ManticoreGeneration.Local,
    resolution=1024
)

print(f"Found {len(velocity_files)} velocity field files")

# Get MCMC step 0 (files are sorted by MCMC number)
velocity_mcmc0 = velocity_files[0]

# Disable caching to download directly
velocity_mcmc0.cache_enabled = False

# Download the file
velocity_mcmc0.download("velocity_mcmc0.hdf5")
print("\nVelocity field downloaded successfully!")
````

## Load and Process the Data

Let's load the density and momentum fields, then calculate the velocity.

````python
# Load the HDF5 file and extract density and momentum
with h5py.File("velocity_mcmc0.hdf5", 'r') as f:
    # Print available datasets
    print("Datasets in velocity field file:")
    print(list(f.keys()))
    print()
    
    # Load density field
    density = f['density'][:]
    
    # Load momentum components
    p0 = f['p0'][:]  # x-component
    p1 = f['p1'][:]  # y-component
    p2 = f['p2'][:]  # z-component
    
print(f"Density field shape: {density.shape}")
print(f"Density range: [{density.min():.3e}, {density.max():.3e}]")
print()
print(f"Momentum field shape: {p0.shape}")
print(f"Momentum range (p0): [{p0.min():.3e}, {p0.max():.3e}]")
````

````python
# Calculate velocity from momentum and density
# v = p / ρ (velocity in km/s)

# Avoid division by zero
epsilon = 1e-10
density_safe = np.where(density > epsilon, density, epsilon)

# Calculate velocity components
v0 = p0 / density_safe  # x-component (km/s)
v1 = p1 / density_safe  # y-component (km/s)
v2 = p2 / density_safe  # z-component (km/s)

# Calculate velocity magnitude
velocity_magnitude = np.sqrt(v0**2 + v1**2 + v2**2)

print(f"Velocity magnitude range: [{velocity_magnitude.min():.2f}, {velocity_magnitude.max():.2f}] km/s")
print(f"Mean velocity magnitude: {velocity_magnitude.mean():.2f} km/s")
````

## Plot Slices Through the Fields

Let's visualize 2D slices through the middle of each field.

````python
# Simulation parameters
L = 1000*0.681  # Box size in Mpc/h
ngrid = density.shape[0]  # 256
cell_size = L / ngrid  # Mpc/h per cell

# Take a slice through the middle of the volume
slice_idx = density.shape[0] // 2

# Convert mass to surface density (mass per unit area)
mass_slice = density[slice_idx, :, :]
surface_density = mass_slice / (cell_size**2)  # h M☉ / Mpc²

# Create the figure
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot surface density slice
im1 = axes[0].imshow(
    np.log10(surface_density + 1e-10),  # Log scale, add small value to avoid log(0)
    cmap='viridis',
    origin='lower',
    interpolation='nearest'
)
axes[0].set_title(f'Surface Density (slice at z idx={slice_idx})', fontsize=14, fontweight='bold')
axes[0].set_xlabel('X [grid cells]')
axes[0].set_ylabel('Y [grid cells]')
cbar1 = plt.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)
cbar1.set_label(r'log₁₀(Σ) [$h$ M$_\odot$ Mpc$^{-2}$]', rotation=270, labelpad=20)

# Plot velocity magnitude slice
im2 = axes[1].imshow(
    velocity_magnitude[slice_idx, :, :],
    cmap='plasma',
    origin='lower',
    interpolation='nearest'
)
axes[1].set_title(f'Velocity Magnitude (slice at z idx={slice_idx})', fontsize=14, fontweight='bold')
axes[1].set_xlabel('X [grid cells]')
axes[1].set_ylabel('Y [grid cells]')
cbar2 = plt.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)
cbar2.set_label('|v| [km/s]', rotation=270, labelpad=20)

plt.tight_layout()
plt.show()
````

# Plotting from the perspective of the observer

## Coordinate Setup

### Simulation Box
- **Density field shape**: 256³ voxels
- **Box size**: L = 681 Mpc/h (comoving)
- **Cell size**: ~3.91 Mpc/h per voxel
- **Redshift**: z = 0

### Coordinate System
- **Field coordinates**: Cartesian equatorial (X, Y, Z)
  - X-axis: points toward RA = 0°, Dec = 0°
  - Y-axis: points toward RA = 90°, Dec = 0°
  - Z-axis: points toward Dec = 90° (North Celestial Pole)
- **Grid units**: indices [0-255] in each dimension
- **Physical units**: 0 to 681 Mpc/h in each dimension

### Observer Position
- **Location**: Center of the box
  - Physical: (500, 500, 500) Mpc/h
  - Grid units: (128, 128, 128)
- **Shell sampling**: Spherical shells centered on observer
  - d1, d2: inner/outer radius in Mpc/h
  - Radial sampling: 20 points averaged through shell thickness

### Projection
- **Map**: HEALPix mollweide projection (nside=128)
- **Orientation**: RA = 0° at center, increasing eastward
- **Quantity**: Overdensity δ = ρ/ρ̄ - 1, shown as sign(δ) × log₁₀(1 + |δ|)

The figure below emulates the Figure 2 from the Manticore-Local paper (note that that figure was made from the simulation dark matter particles directly and thus could be made at much higher resolution).

````python
import numpy as np
import healpy as hp
from scipy.ndimage import map_coordinates

def plot_shell(density, observer_pos, d1, d2, L=681, nside=128):
    """
    Plot a spherical shell surface density from a 3D density field on HEALPix projection.
    
    Parameters:
    -----------
    density : array (256,256,256)
        3D mass field in M☉/h per cell
    observer_pos : tuple or array
        Observer position in grid units (ix, iy, iz)
    d1, d2 : float
        Inner and outer radius of shell in Mpc/h
    L : float
        Box size in Mpc/h
    nside : int
        HEALPix resolution parameter
    """
    # Grid parameters
    ngrid = density.shape[0]
    cell_size = L / ngrid  # Mpc/h per cell
    
    # Get HEALPix pixel directions (theta, phi)
    npix = hp.nside2npix(nside)
    theta, phi = hp.pix2ang(nside, np.arange(npix))
    
    # Convert to Cartesian unit vectors
    nx = np.sin(theta) * np.cos(phi)
    ny = np.sin(theta) * np.sin(phi)
    nz = np.cos(theta)
    
    # Sample radii in the shell
    n_samples = 20
    radii = np.linspace(d1, d2, n_samples)
    dr = (d2 - d1) / (n_samples - 1) if n_samples > 1 else (d2 - d1)
    
    surface_density = np.zeros(npix)
    
    # For each pixel, integrate along the line of sight through the shell
    for i, r in enumerate(radii):
        # Position in grid units
        pos_x = observer_pos[0] + r * nx / cell_size
        pos_y = observer_pos[1] + r * ny / cell_size
        pos_z = observer_pos[2] + r * nz / cell_size
        
        # Stack coordinates for map_coordinates
        coords = np.array([pos_x, pos_y, pos_z])
        
        # Sample density field (with boundary handling)
        # density is in M☉/h per cell, convert to volume density
        sampled = map_coordinates(density, coords, order=1, 
                                   mode='constant', cval=0)
        
        # Convert from mass per cell to volume density and integrate
        # Volume density: M☉/h / (Mpc/h)³
        volume_density = sampled / (cell_size**3)
        
        # Integrate along line of sight: Σ = ∫ ρ dr
        surface_density += volume_density * dr
    
    # surface_density is now in M☉/h per (Mpc/h)²
    # Take log for visualization
    log_surface_density = np.log10(surface_density + 1e-10)
    
    # Plot
    hp.projview(log_surface_density, 
                title=f'Shell: {d1}-{d2} Mpc/h (Surface Density)', 
                unit=r'log$_{10}$(Σ) [M$_\odot$h Mpc$^{-2}$]',
                cmap='viridis',
                graticule=True,
                graticule_labels=True,
                coord="C",
                phi_convention="clockwise",
                rot=[0,0,0],
                flip="geo",
                min=11, max=13)

# Example usage:
observer = (128, 128, 128)  # Center of box
plot_shell(density, observer, d1=95, d2=125, L=681, nside=128)
````
