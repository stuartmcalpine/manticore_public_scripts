## Import Required Libraries

````python
import numpy as np
import matplotlib.pyplot as plt
import h5py
from manticore_data import ManticoreDataLoader, ManticoreGeneration

# Set up matplotlib for better-looking plots
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
````

## Initialize the Data Loader

The `ManticoreDataLoader` connects to the S3 data repository and discovers available data products.

````python
# Initialize the loader
loader = ManticoreDataLoader()

# Check available products
#print("Available products:")
#products = loader.get_products(ManticoreGeneration.Local)
#for product in products:
#    print(f"  - {product}")
````

## Download Velocity and Density Field for MCMC Step 0

The velocity field file contains both density and momentum data. We'll download it without caching to a local file. There are a total of 80 field files (one for each Manticore-Local indepedent realization).

**File structure:**
- `density`: Density field
- `p0`, `p1`, `p2`: Momentum components (x, y, z)
- To get velocity: divide momentum by density (velocity in km/s)

````python
# Get velocity field files for resolution 1024
velocity_files = loader.get_velocity_field(
    ManticoreGeneration.Local,
    resolution=1024
)

print(f"Found {len(velocity_files)} velocity field files")

# Get MCMC step 0 (files are sorted by MCMC number)
velocity_mcmc0 = velocity_files[0]

# Disable caching to download directly
velocity_mcmc0.cache_enabled = False

# Download the file
velocity_mcmc0.download("velocity_mcmc0.hdf5")
print("\nVelocity field downloaded successfully!")
````

## Load and Process the Data

Let's load the density and momentum fields, then calculate the velocity.

````python
# Load the HDF5 file and extract density and momentum
with h5py.File("velocity_mcmc0.hdf5", 'r') as f:
    # Print available datasets
    print("Datasets in velocity field file:")
    print(list(f.keys()))
    print()
    
    # Load density field
    density = f['density'][:]
    
    # Load momentum components
    p0 = f['p0'][:]  # x-component
    p1 = f['p1'][:]  # y-component
    p2 = f['p2'][:]  # z-component
    
print(f"Density field shape: {density.shape}")
print(f"Density range: [{density.min():.3e}, {density.max():.3e}]")
print()
print(f"Momentum field shape: {p0.shape}")
print(f"Momentum range (p0): [{p0.min():.3e}, {p0.max():.3e}]")
````

````python
# Calculate velocity from momentum and density
# v = p / ρ (velocity in km/s)

# Avoid division by zero
epsilon = 1e-10
density_safe = np.where(density > epsilon, density, epsilon)

# Calculate velocity components
v0 = p0 / density_safe  # x-component (km/s)
v1 = p1 / density_safe  # y-component (km/s)
v2 = p2 / density_safe  # z-component (km/s)

# Calculate velocity magnitude
velocity_magnitude = np.sqrt(v0**2 + v1**2 + v2**2)

print(f"Velocity magnitude range: [{velocity_magnitude.min():.2f}, {velocity_magnitude.max():.2f}] km/s")
print(f"Mean velocity magnitude: {velocity_magnitude.mean():.2f} km/s")
````

## Plot Slices Through the Fields

Let's visualize 2D slices through the middle of each field.

````python
# Take a slice through the middle of the volume
slice_idx = density.shape[0] // 2

# Create the figure
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot density slice
im1 = axes[0].imshow(
    np.log10(density[slice_idx, :, :] + 1e-10),  # Log scale, add small value to avoid log(0)
    cmap='viridis',
    origin='lower',
    interpolation='nearest'
)
axes[0].set_title(f'Density Field (slice at z={slice_idx})', fontsize=14, fontweight='bold')
axes[0].set_xlabel('X [grid cells]')
axes[0].set_ylabel('Y [grid cells]')
cbar1 = plt.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)
cbar1.set_label('log₁₀(ρ)', rotation=270, labelpad=20)

# Plot velocity magnitude slice
im2 = axes[1].imshow(
    velocity_magnitude[slice_idx, :, :],
    cmap='plasma',
    origin='lower',
    interpolation='nearest'
)
axes[1].set_title(f'Velocity Magnitude (slice at z idx={slice_idx})', fontsize=14, fontweight='bold')
axes[1].set_xlabel('X [grid cells]')
axes[1].set_ylabel('Y [grid cells]')
cbar2 = plt.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)
cbar2.set_label('|v| [km/s]', rotation=270, labelpad=20)

plt.tight_layout()
plt.show()
````

## Summary Statistics

````python
print("=" * 60)
print("MCMC Step 0 - Field Statistics")
print("=" * 60)
print(f"\nDensity Field:")
print(f"  Shape: {density.shape}")
print(f"  Mean: {density.mean():.3e}")
print(f"  Std: {density.std():.3e}")
print(f"  Min: {density.min():.3e}")
print(f"  Max: {density.max():.3e}")

print(f"\nMomentum Field (p0):")
print(f"  Shape: {p0.shape}")
print(f"  Mean: {p0.mean():.3e}")
print(f"  Std: {p0.std():.3e}")
print(f"  Min: {p0.min():.3e}")
print(f"  Max: {p0.max():.3e}")

print(f"\nVelocity Field (derived from p/ρ):")
print(f"  Shape: {velocity_magnitude.shape}")
print(f"  Mean magnitude: {velocity_magnitude.mean():.2f} km/s")
print(f"  Std magnitude: {velocity_magnitude.std():.2f} km/s")
print(f"  Min magnitude: {velocity_magnitude.min():.2f} km/s")
print(f"  Max magnitude: {velocity_magnitude.max():.2f} km/s")
print("\nVelocity components (mean):")
print(f"  v_x: {v0.mean():.2f} km/s")
print(f"  v_y: {v1.mean():.2f} km/s")
print(f"  v_z: {v2.mean():.2f} km/s")
print("=" * 60)
````
