# Manticore Data Tutorial

This notebook demonstrates how to use the `manticore_data` package to download and visualize simulation data from the Manticore project.

````python
import numpy as np
import matplotlib.pyplot as plt
import h5py
from manticore_data import ManticoreDataLoader, ManticoreGeneration

# Set up matplotlib for better-looking plots
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
````

## Initialize the Data Loader

The `ManticoreDataLoader` connects to the S3 data repository and discovers available data products.

````python
# Initialize the loader
loader = ManticoreDataLoader()
````

## Download a Velocity Field

Download the velocity field for MCMC step 0 (one of 80 posterior realizations). The file contains a 256³ grid with:

- `density`: mass field [M☉/h per cell]
- `p0`, `p1`, `p2`: momentum components (x, y, z) [(M☉/h)(km/s)]
- Velocity is obtained by `v = p / density` [km/s]

`h = 0.681`

````python
# Get velocity field files for resolution 1024
velocity_files = loader.get_velocity_field(
    ManticoreGeneration.Local,
    resolution=1024
)

print(f"Found {len(velocity_files)} velocity field files")

# Get MCMC step 0 (files are sorted by MCMC number)
velocity_mcmc0 = velocity_files[0]

# Disable caching to download directly
velocity_mcmc0.cache_enabled = False

# Download the file
velocity_mcmc0.download("velocity_mcmc0.hdf5")
print("\nVelocity field downloaded successfully!")
````

## Load and Process the Data

Let's load the density and momentum fields, then calculate the velocity.

````python
# Load density and momentum fields
with h5py.File("velocity_mcmc0.hdf5", 'r') as f:
    density = f['density'][:]    # [M☉/h per cell]
    p0 = f['p0'][:]              # x-momentum [(M☉/h)(km/s)]
    p1 = f['p1'][:]              # y-momentum
    p2 = f['p2'][:]              # z-momentum

print(f"Field shape: {density.shape}")

# Calculate velocity: v = p / ρ
epsilon = 1e-10
density_safe = np.where(density > epsilon, density, epsilon)
v0 = p0 / density_safe
v1 = p1 / density_safe
v2 = p2 / density_safe
velocity_magnitude = np.sqrt(v0**2 + v1**2 + v2**2)
````

````python
# Simulation parameters
L = 1000*0.681  # Box size in Mpc/h
ngrid = density.shape[0]  # 256
cell_size = L / ngrid  # Mpc/h per cell

# Take a slice through the middle of the volume
slice_idx = density.shape[0] // 2

# Convert mass to surface density (mass per unit area)
mass_slice = density[slice_idx, :, :]
surface_density = mass_slice / (cell_size**2)  # h M☉ / Mpc²

# Create the figure
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot surface density slice
im1 = axes[0].imshow(
    np.log10(surface_density + 1e-10),  # Log scale, add small value to avoid log(0)
    cmap='viridis',
    origin='lower',
    interpolation='nearest'
)
axes[0].set_title(f'Surface Density (slice at z idx={slice_idx})', fontsize=14, fontweight='bold')
axes[0].set_xlabel('X [grid cells]')
axes[0].set_ylabel('Y [grid cells]')
cbar1 = plt.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)
cbar1.set_label(r'log₁₀(Σ) [$h$ M$_\odot$ Mpc$^{-2}$]', rotation=270, labelpad=20)

# Plot velocity magnitude slice
im2 = axes[1].imshow(
    velocity_magnitude[slice_idx, :, :],
    cmap='plasma',
    origin='lower',
    interpolation='nearest'
)
axes[1].set_title(f'Velocity Magnitude (slice at z idx={slice_idx})', fontsize=14, fontweight='bold')
axes[1].set_xlabel('X [grid cells]')
axes[1].set_ylabel('Y [grid cells]')
cbar2 = plt.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)
cbar2.set_label('|v| [km/s]', rotation=270, labelpad=20)

plt.tight_layout()
plt.show()
````

## Sky Map from the Observer's Perspective

Project the density field onto a HEALPix Mollweide sky map as seen from the observer at the centre of the box, integrating through a configurable spherical shell.

- **Box**: 681 Mpc/h comoving, 256³ voxels (~2.66 Mpc/h per cell)
- **Observer**: centre of the box at grid index (128, 128, 128)
- **Coordinates**: Cartesian equatorial — X toward (RA=0°, Dec=0°), Y toward (RA=90°, Dec=0°), Z toward Dec=90°
- **Projection**: HEALPix Mollweide, RA increasing eastward

The figure below replicates the style of Figure 2 from McAlpine et al. (2025).

````python
import healpy as hp
from scipy.ndimage import map_coordinates

def plot_shell(density, observer_pos, d1, d2, L=681, nside=128):
    """
    Project a spherical shell of the density field onto a HEALPix sky map.

    Parameters
    ----------
    density      : (256, 256, 256) array — mass field in M☉/h per cell
    observer_pos : (ix, iy, iz) — observer position in grid units
    d1, d2       : inner/outer shell radius in Mpc/h
    L            : box size in Mpc/h
    nside        : HEALPix resolution
    """
    ngrid = density.shape[0]
    cell_size = L / ngrid

    npix = hp.nside2npix(nside)
    theta, phi = hp.pix2ang(nside, np.arange(npix))

    nx = np.sin(theta) * np.cos(phi)
    ny = np.sin(theta) * np.sin(phi)
    nz = np.cos(theta)

    n_samples = 20
    radii = np.linspace(d1, d2, n_samples)
    dr = (d2 - d1) / (n_samples - 1) if n_samples > 1 else (d2 - d1)

    surface_density = np.zeros(npix)
    for r in radii:
        coords = np.array([
            observer_pos[0] + r * nx / cell_size,
            observer_pos[1] + r * ny / cell_size,
            observer_pos[2] + r * nz / cell_size,
        ])
        sampled = map_coordinates(density, coords, order=1, mode='constant', cval=0)
        surface_density += (sampled / cell_size**3) * dr

    hp.projview(
        np.log10(surface_density + 1e-10),
        title=f'Shell: {d1:.0f}–{d2:.0f} Mpc/h',
        unit=r'log$_{10}$(\Sigma) [M$_\odot$h Mpc$^{-2}$]',
        cmap='viridis',
        graticule=True,
        graticule_labels=True,
        coord='C',
        phi_convention='clockwise',
        rot=[0, 0, 0],
        flip='geo',
        min=11, max=13,
    )


observer = (128, 128, 128)  # centre of box
plot_shell(density, observer, d1=95*0.681, d2=125*0.681, L=681, nside=128)
````
